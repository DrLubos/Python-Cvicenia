### Priklad2.

Nakreslite graf funkcie

$f:y=x\mathrm e^{-x^2}$

pre $x\in \langle -3, 5 \rangle.$

```python
x = np.linspace(-3,5,100)
y = x*np.exp(-x**2)
plt.plot(x,y, color='red');
plt.grid()
plt.title('$f:y=x.e^{-x^2}$')
```

### Priklad 3.

Nakreeslite do jedneho obrazka grafy funkcii

$f:y=\sin{x}$

$g:y=\cos{x}$

pre $x\in \langle 0, 2\pi \rangle.$

```python
f = lambda x:np.sin(x)
g = lambda x:np.cos(x)
```

```python
x = np.linspace(0, 6.28, 100)
y1 = f(x)
y2 = g(x)
plt.plot(x, y1, label = 'y=sin(x)');
plt.plot(x, y2, label = 'y=cos(x)');
plt.grid()
plt.legend(loc=1);
plt.xlabel('os x');
plt.ylabel('os y');
```

### Priklad 4.

Nakreslite graf funkcie dvoch premennych pomocou vrstevnic

$z=f(x,y)=(x^2+y^2)\sin{\frac{1}{xy}}$

pre $x\in\langle -2,2\rangle,y\in\langle -2.5, 2.5\rangle.$

```python
x = np.linspace(-2, 2, 100)
y = np.linspace(-2.5, 2.5, 100)
X, Y = plt.np.meshgrid(x, y)
Z = (X**2 + Y**2) * np.sin(1/(X*Y))
plt.contour(X, Y, Z);
plt.colorbar();
```

```python
x = np.linspace(-2, 2, 100)
y = np.linspace(-2.5, 2.5, 100)
X, Y = plt.np.meshgrid(x, y)
Z = (X**2 + Y**2) * np.sin(1/(X*Y))
plt.contourf(X, Y, Z);
plt.colorbar();
plt.savefig('fnc.png')
```

```python
def kresli(predpis, od, do):
    fnc = eval('lambda x: ' + predpis)
    x = np.linspace(od, do, 100)
    y = fnc(x)
    plt.plot(x, y);
    plt.grid()
```
